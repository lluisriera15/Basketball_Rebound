```python
# LOAD GENERAL MODULES

import pandas as pd
import numpy as np
import os
import csv
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')  # Cambiar el backend a TkAgg
from pathlib import Path

# LOAD MODULES VIDEO  AI

import cv2
import ultralytics
from ultralytics import YOLO

# LOAD GENERAL FROM OTHER NOTEBOOKS

import nbimporter
import tracking_video_module
import rebouding_identification_module
```

```python
# PROCESS ONE VIDEO AND ADD FINAL DATA TO THE 

def VIDEO_TO_REBOUND (video_path):
    model = YOLO('models/bestbasket.pt')
    model_pose = YOLO('models/bestbasket-pose.pt')
    
    left_court_2D = pd.read_csv('csv/points2D_left_court.csv', skiprows=1, header=None).apply(tuple, axis=1).tolist()
    right_court_2D = pd.read_csv('csv/points2D_right_court.csv', skiprows=1, header=None).apply(tuple, axis=1).tolist()
    
    results = tracking_video_module.PROCESS_VIDEO (video_path, model)
    results_pose = tracking_video_module.PROCESS_VIDEO_POSE (video_path, model_pose)
    df = tracking_video_module.EXTRACT_RESULTS (results, video_path)
    df_pose = tracking_video_module.EXTRACT_RESULTS_POSE(results_pose, video_path)
    df_pose = tracking_video_module.CLEAN_DATA_POSE (df_pose)
    
    df = rebouding_identification_module.UNIQUE_BALLS(df)
    df = rebouding_identification_module.INTERPOLATE_BALL_HOOP (df)
    df = rebouding_identification_module.IDENTIFY_POSSESSION(df)
    df_pose =  rebouding_identification_module.INTERPOLATE_ALL_VALUES (df_pose)
    df_pose = rebouding_identification_module.HOMOGRAPHY_MATRIX (df_pose, left_court_2D, right_court_2D)
    df = rebouding_identification_module.CALCULATE_HOOP_PROJECTION (df, df_pose)
    df = rebouding_identification_module.POINTS_FLOOR(df)
    df = rebouding_identification_module.APPLY_HOMOGRAPHY (df, df_pose)
    df_shot = rebouding_identification_module.DETECT_SHOT (df, video_path)
    df_shot_points = rebouding_identification_module.EXTRACT_SHOT_POINTS(df_shot)
    
    return (df_shot_points)
```

```python
video_path = 'videos/videotest8.mp4'
df_shot_points = VIDEO_TO_REBOUND (video_path)
df_shot_points
```

```python
# PROCESS ONE FOLDER FULL OF VIDEOS

def FOLDER_TO_REBOUND (folder_path, csv_file):
    
    df_rebound = pd.DataFrame(columns=['video_name', 'x_2D_shot', 'y_2D_shot', 'x_2D_rebound', 'y_2D_rebound'])
    df_rebound.to_csv('/csv/off_reb.csv', index=False)
    i=1

    for filename in os.listdir(folder_path):
        print(f"Analizing file number {i} with the name: {filename}")
        video_path = os.path.join(folder_path, filename)
        name_file = Path(video_path).stem
        df_rebound = VIDEO_TO_REBOUND (video_path)
        df_rebound.insert(0, 'video_name', name_file)
        df_rebound.to_csv(csv_file, mode='a', header=False, index=False)
        os.remove(video_path)
        i += 1
    return (df_rebound)
```

```python
folder_path = 'videos_rebound/'
csv_file = 'csv/videos_prova.csv'
FOLDER_TO_REBOUND (folder_path, csv_file)
```
